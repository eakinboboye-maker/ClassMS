# EEE355 Mock Exam Template

```python
import os, sys
sys.path.append(os.path.abspath("../_shared"))
from classlite_jupyter import ClassLiteConfig, ClassLiteNotebookApp

config = ClassLiteConfig(
    api_base="http://localhost:8000",
    assessment_id=1,
    attendance_session_id=1,
    lesson_slug="mock_exam_01",
    question_ids={"mcq_1": 1, "fill_gap_1": 2, "essay_1": 3},
)
app = ClassLiteNotebookApp(config)
```

## Login

```python
app.render_login_widget()
```

## Start Attempt

```python
app.render_attempt_controls()
```

## MCQ

```python
app.render_mcq_single(
    question_id=config.question_ids["mcq_1"],
    prompt="Q1. Which quantity is a vector?",
    options=[("a", "mass"), ("b", "velocity"), ("c", "temperature")],
)
```

## Fill Gap

```python
app.render_fill_gap(
    question_id=config.question_ids["fill_gap_1"],
    prompt="Q2. Energy cannot be [GAP1] or [GAP2].",
    gap_keys=["GAP1", "GAP2"],
)
```

## Essay

```python
app.render_essay(
    question_id=config.question_ids["essay_1"],
    prompt="Q3. Explain the difference between Moore and Mealy machines.",
)
```
