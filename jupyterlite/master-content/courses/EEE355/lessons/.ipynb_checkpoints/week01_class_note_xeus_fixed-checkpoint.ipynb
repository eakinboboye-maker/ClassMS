## Main note

Digital abstraction means representing information using **discrete values**.


```python
def analog_to_digital_example(value, threshold=0.5):
    return 1 if value >= threshold else 0

samples = [0.1, 0.3, 0.49, 0.5, 0.7, 0.95]
[(x, analog_to_digital_example(x)) for x in samples]
```
